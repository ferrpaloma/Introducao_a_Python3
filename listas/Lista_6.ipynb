## Problema 1: Informações em um texto

Processe um arquivo texto chamado **informacoes.txt**. Este arquivo contém a informações de pessoas contendo seus nomes, idade, sexo e cidade aonde mora, nesta mesma ordem. Peça que o usario digite o nome de algum deles e imprima na saída padrão uma mensagem

### Entrada

A entrada consiste do nome do arquivo e no nome da pessoa

### Saída

A saída consiste uma mensagem personalizada para cada nome digitado, seguindo o padrão a baixo

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| informacoes.txt<br>Paulo | Paulo e um(a) homem de 22 anos morador(a) de Nova Iguacu |
| informacoes.txt<br>Douglas | Douglas e um(a) homem de 18 anos morador(a) de Duque de Caxias |
| informacoes.txt<br>Jennifer | Jennifer e um(a) mulher de 22 anos morador(a) de Paracambi |
| informacoes.txt<br>Paloma | Paloma e um(a) mulher de 19 anos morador(a) de Paty do Alferes |

```
# SP
def carregar_informacoes(nome_arquivo):
  informacoes = dict()
  arquivo = open(caminho+nome_arquivo,"r")
  for linha in arquivo:
    linha = linha.strip()
    linha = linha.split(sep= " ",maxsplit =3)
    informacoes[linha[0]] = {"idade":linha[1],"sexo":linha[2],"cidade":linha[3]}
  arquivo.close()
  return informacoes
```

```
# PP
arquivo = input("Nome do arquivo + extensao: ")
informacoes = carregar_informacoes(arquivo)
#print(informacoes)
pessoa = input("Digite o nome da pessoa:")
print("{} e um(a) {} de {} anos morador(a) de {}".format(pessoa,informacoes[pessoa]["sexo"],informacoes[pessoa]["idade"],informacoes[pessoa]["cidade"]))
```
