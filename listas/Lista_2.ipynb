|:-|:-|

## (DESAFIO) Problema 7: Tipos de triângulos

Leia 3 valores de ponto flutuante, **A**, **B** e **C**, em ordem decrescente de valores, de modo que o lado **A** representa o maior dos três lados. Crie uma função para determina o tipo de triângulo que estes três lados formam, com base nos seguintes casos, sempre escrevendo uma mensagem adequada:

* Se A ≥ B + C, apresente a mensagem: NAO FORMA TRIANGULO;
* Se A² = B² + C² , apresente a mensagem: TRIANGULO RETANGULO;
* Se A² > B² + C² , apresente a mensagem: TRIANGULO OBTUSANGULO;
* Se A² < B² + C² , apresente a mensagem: TRIANGULO ACUTANGULO;
* Se os três lados forem iguais, apresente a mensagem: TRIANGULO
EQUILATERO;
* Se apenas dois dos lados forem iguais, apresente a mensagem: TRIANGULO
ISOSCELES.

### Entrada

A entrada contem três valores de ponto flutuante de dupla precisão: A (0 < A), B (0 < B) e C (0 < C).

### Saída

Imprima todas as classificações do triângulo especificado na entrada.

|:-|:-|
|7.0<br>7.0<br>5.0 | TRIANGULO ACUTANGULO<br>TRIANGULOS ISOSCELES|
|10.0<br>6.0<br>6.0| TRIANGULO OBTUSANGULO<br>TRIANGULO ISOSCELES|
|7.0<br>5.0<br>2.0 | NAO FORMA TRIANGULO |
|10.0<br>8.0<br>6.0| TRIANGULO RETANGULO|
