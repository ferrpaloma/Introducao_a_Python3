## Problema 1: Informações em um texto

Processe um arquivo texto chamado **informacoes.txt**. Este arquivo contém a informações de pessoas contendo seus nomes, idade, sexo e cidade aonde mora, nesta mesma ordem. Peça que o usario digite o nome de algum deles e imprima na saída padrão uma mensagem

### Entrada

A entrada consiste do nome do arquivo e no nome da pessoa

### Saída

A saída consiste uma mensagem personalizada para cada nome digitado, seguindo o padrão a baixo

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| informacoes.txt<br>Paulo | Paulo e um(a) homem de 22 anos morador(a) de Nova Iguacu |
| informacoes.txt<br>Douglas | Douglas e um(a) homem de 18 anos morador(a) de Duque de Caxias |
| informacoes.txt<br>Jennifer | Jennifer e um(a) mulher de 22 anos morador(a) de Paracambi |
| informacoes.txt<br>Paloma | Paloma e um(a) mulher de 19 anos morador(a) de Paty do Alferes |

```
# SP
def carregar_informacoes(nome_arquivo):
  informacoes = dict()
  arquivo = open(caminho+nome_arquivo,"r")
  for linha in arquivo:
    linha = linha.strip()
    linha = linha.split(sep= " ",maxsplit =3)
    informacoes[linha[0]] = {"idade":linha[1],"sexo":linha[2],"cidade":linha[3]}
  arquivo.close()
  return informacoes
```

```
# PP
arquivo = input("Nome do arquivo + extensao: ")
informacoes = carregar_informacoes(arquivo)
#print(informacoes)
pessoa = input("Digite o nome da pessoa:")
print("{} e um(a) {} de {} anos morador(a) de {}".format(pessoa,informacoes[pessoa]["sexo"],informacoes[pessoa]["idade"],informacoes[pessoa]["cidade"]))
```

## Problema 2: Nomes e idades

Faça um programa que crie um dicionário de nomes de pessoas e idades, onde o nome de cada pessoa é a chave associada ao valor idade (do tipo inteiro). Peça ao usuário para digitar nomes e idades e preencha o dicionário com as informações dadas, até que uma string vazia seja digitada (que não deve ser colocada no dicionário). No fim exiba nome e idade ordenada pelos nomes em ordem alfabética, ou seja, ordenado pela chave em ordem alfabética.

**OBS:** dict.keys() é um objeto iteravél, ou seja, pode ser usada no for.

**OBS:** sorted() é uma função que ordena um iteravél.

### Entrada

A entrada consiste em nomes, strings, e idades, inteiros.

### Saída

A saída consiste no dicionario ordenado alfabeticamente pelas chaves.

```
# SP
def receber_nome_idade():
  pessoas = {}
  nome_idade = input("Digite seu nome e idade (Paulo 22): ") 
  while nome_idade != "":
    nome, idade = nome_idade.split(sep = " ")
    pessoas[nome] = int(idade)
    nome_idade = input("Digite seu nome e idade (Paulo 22): ") 
  return pessoas

def imprimir_nomes_ordenados(pessoas):
  pessoas_ordenados = {}
  for chave in sorted(pessoas.keys()):
    pessoas_ordenados[chave] = pessoas[chave]
  print(pessoas_ordenados)
  return None
```

```
# PP
pessoas = receber_nome_idade()
imprimir_nomes_ordenados(pessoas)
```

## Problema 3: Letras em uma palavra

Escreva um programa que peça ao usuário para digitar uma cadeia de caracteres via entrada padrão, e que retorna um dicionário cujas chaves são as letras do alfabeto em ordem alfabética que ocorrem na cadeia de caracteres, e seus valores são o número de vezes que cada letra ocorre.

**OBS:** Se a chave não existir atribua um valor inicial 0 (zero)

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| fundamentos de programacao | {a: 4, c: 1, d: 2, e: 2, f: 1, g: 1, m: 2, n: 2, o: 3, p: 1, r: 2, s: 1, t: 1, u: 1} |
| palavra | {a: 3, l: 1, p: 1, r: 1, v: 1} |
| escola de inverno | {a: 1, c: 1, d: 1, e: 3, i: 1, l: 1, n: 2, o: 2, r: 1, s: 1, v: 1} |


```
def criar_dicionario(palavra):
  letras = {}
  for letra in palavra:
    if letra == " ":
      continue
    elif letra not in letras.keys():
      letras[letra] = 1
    else:
      letras[letra] += 1
  return letras

def letras_ordenadas(letras):
  letras_ordenados = {}
  for chave in sorted(letras.keys()):
    letras_ordenados[chave] = letras[chave]
  print(letras_ordenados)
  return None
```

```
# PP
palavra = input("Digite uma palavra ou uma frase: ")
letras = criar_dicionario(palavra)
letras_ordenadas(letras)
```

```
# PP
palavra = input("Digite uma palavra ou uma frase: ")
letras = criar_dicionario(palavra)
letras_ordenadas(letras)
```

## Problema 4: Temperatura

Processe um arquivo texto chamado **temperaturas.txt**. Este arquivo contém a informações de cidades contendo seus nomes e temperatura média em um dia, separados por uma virgula (,). Retorne a cidade com temperatura mais quente e a mais fria do dia.

**OBS:** Caso for usar a funcao max/min em um dicionario: max/min(dicionario,key=dicionario.get)

### Entrada

A entrada consiste do nome do arquivo

### Saída

A saída consiste em duas strings, o nome da cidade mais quente e a mais fria do dia

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| temperaturas.txt | mais quente = Itaperuna <br> mais fria = Mangaratiba |

```
def abrir_arquivo(arquivo):
  dados = open(caminho+arquivo+".txt","r")
  temperaturas = dict()
  for linha in dados:
    linha = linha.strip()
    linha = linha.split(sep = ",")
    temperaturas[linha[0]] = int(linha[1])
  return temperaturas
```

```
# PP
arquivo = input("Digite o nome do arquivo sem extensao: ")
temperaturas = abrir_arquivo(arquivo)
cidade_quente = max(temperaturas,key=temperaturas.get)
cidade_fria = min(temperaturas,key=temperaturas.get)
print("mais quente =",cidade_quente,"\nmais fria =",cidade_fria)
```

## Problema 5:

Processe um arquivo texto que contém um texto e crie um dicionario onde as chaves são as letras e seu valor é quantas vezes a letra se repete no texto. Imprima a letra que mais se repete e seu valor

**OBS:** Caso for usar a funcao max/min em um dicionario: max/min(dicionario,key=dicionario.get)

### Entrada

A entrada consiste do nome do arquivo

### Saída

A saída consiste em uma string e um inteiro.

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| retrato.txt | o 13 |
| amavisse.txt | e 19 |

```
# SP
def abrir_arquivo(arquivo):
  dados = open(caminho+arquivo+".txt","r")
  letras = {}
  for linha in dados:
    linha = linha.strip()
    for letra in linha:
      if letra == " ":
        continue
      elif letra not in letras.keys():
        letras[letra] = 1
      else:
        letras[letra] += 1
  return letras
```

```
# PP
arquivo = input("Digite o nome do arquivo sem extensao: ")
letras = abrir_arquivo(arquivo)
letra_mais_repetida = max(letras,key=letras.get)
print(letra_mais_repetida,letras[letra_mais_repetida])
```
