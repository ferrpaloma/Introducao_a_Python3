## Problema 2: Peso ideal

Crie uma função chamada **peso_ideal** que recebe a altura e o sexo de uma pessoa e retorna o peso ideal utilizando as seguintes fórmulas (onde h corresponde à altura):

  * Homens: (72.7*h)-58
  * Mulheres: (62.1*h)-44.7

Todas as entradas tem que ser fornecidas pelo usuário

### Entrada

A entrada contém o sexo como str e a altura como float com duas casas decimais

### Saída 

Imprima o peso ideal que a função retorna com três casas decimais aṕos o ponto, conforme o exemplo a baixo


| **Exemplos de Entrada** | **Exemplos de Saída** |
|:--|:--|
| masculino<br>1.69 | O seu peso ideal é 64.863 kg |
| feminino<br>1.93 | O seu peso ideal é 75.153 kg |

```
# SP
def peso_ideal(altura,sexo):
  if sexo == "masculino":
    peso = (72.7*altura) - 58
  elif sexo == "feminino":
    peso = (62.1*altura) - 44.7
  else:
    peso = 0
  return peso
```

```
# PP
sexo = input("Digite o sexo da pessoa: ") 
altura = float(input("Digite a altura da pessoa: "))
peso = peso_ideal(altura,sexo)
print("Seu peso ideal e: %.3f" % (peso) )
```

## Problema 3: Aprovação

Crie uma função chamada **situacao** que receba as duas notas de um aluno, calcule a média e retorne se ele esta aprovado, recuperação ou reprovado. Assuma que cada nota pode ir de 0 até 10.0, sempre com uma casa decimal. A primeira prova tem peso 1 e a segunda prova peso 2. A situação do aluno segue as seguintes regras:

* Se **media** $\geq$ 6.0, apresente a mensagem: Aluno aprovado
* Se 4.0 $\leq$ **media** $<$ 6.0, apresente a mensagem: Aluno de recuperação
* Se **media** $<$ 4.0, apresente a mensagem: Aluno reprovado

Todas as entradas tem que ser fornecidas pelo usuário

### Entrada

A entrada contem dois valores float com uma casa decimal cada um, que representam as notas do aluno

### Saída

Imprima a situação do aluno que a função retorna

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| 8.5<br>3.5 | Aluno em recuperação |
| 3.7<br>8.3 | Aluno aprovado |
| 5.3<br>3.2 | Aluno reprovado |

```
# SP
def situacao(nota1,nota2):
  media = (nota1*1+nota2*2)/3
  if media >= 6.0:
    print("Aluno aprovado")
  elif media < 6.0 and media >= 4.0:
    print("Aluno em recuperacao")
  elif media < 4.0:
    print("Aluno reprovado")
  return None
```

```
# PP
nota1 = float(input("Digite a nota da primeira prova: "))
nota2 = float(input("Digite a nota da segunda prova: "))
situacao(nota1,nota2)
```

## Problema 4: Categoria

Crie uma função chamada **categoria** que recebe a idade de um nadador e o classifica em uma das seguintes categorias:

| Categoria | Idade |
|:-|:-|
| Infantil A | 5 a 7 anos |
| Infantil B | 8 a 10 anos |
| Juvenil A | 11 a 13 anos |
| Juvenil B | 14 a 17 anos |
| Sênior | maiores de 18 anos |

Todas as entradas tem que ser fornecidas pelo usuário

### Entrada

### Saída

Imprima a classificação que a função retorna

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| 10 | Infantil B |
| 16 | Juvenil B |

```
# SP
def categoria(age):
  if age >= 5 and age <= 7:
    return "Infantil A"
  elif age >= 8 and age <= 10:
    return "Infatil B"
  elif age >= 11 and age <= 13:
    return "Juvenil A"
  elif age >= 14 and age <= 17:
    return "Juvenil B"
  elif age >= 18:
    return "Senior"
  else:
    return "Nao percetence a nenhuma categoria"
```

| **Especificação** | Código | Preço |
|:-|:-|:-|
| Cachorro Quente | 100 | 5.00 |
| Bauru Simples | 101| 20.00 |
| Bauru com Ovo | 102| 23.50 |

| Guaracamp | 106 | 1.50 |

Todas as entradas tem que ser fornecidas pelo usuário

### Entrada

A entrada contém dois valores inteiros sendo o código e quantidade de lanches

### Saída

Imprima, com duas casas decimais, o resultado retornado pela função

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
|104<br>5|R\$ 33.50|

## Problema 6: IMC

Crie uma função chamada **imc** que recebe o peso e altura de uma pessoa e retorna a classificação da pessoa. 

Calculamos o IMC com a função:

* IMC = Peso/Altura²

Já a classificação:

| IMC | Classificação |
|:-|:-|
|$<$ 18.5|Abaixo do Peso|
|18.6 - 24.9|Saudável|
|25.0 - 29.9|Peso em excesso|
|30.0 - 34.9|Obesidade Grau I|
|35.0 - 39.9|Obesidade Grau II (severa)|
|$\geq$ 40.0|Obesidade Grau III (mórbida)|

Todas as entradas tem que ser fornecidas pelo usuário

### Entrada

A entrada contém dois floats, o peso com 3 casas decimais e altura com 2 casas decimais

### Saída

Imprima a classificação do IMC retornado pela função

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
|55.257<br>1.69|Saudável|
|87.462<br>1.53|Obesidade Grau II (severa)|
|93.137<br>1.89|Peso em excesso|



```

# SP
def imc(peso,altura):
  indice = peso/altura**2
  if indice < 18.5:
    print("Abaixo do Peso")
  elif indice >= 18.5  and indice < 25.0:
    print("Saudavel")
  elif indice >= 25.0  and indice < 30.0:
    print("Peso em excesso")
  elif indice >= 30.0  and indice < 35.0:
    print("Obesidade Grau I")
  elif indice >= 35.0  and indice < 40.0:
    print("Obesidade Grau II (severa)")
  elif indice >= 40.0:
    print("Obesidade Grau III (morbida)")
  return None
```

## (DESAFIO) Problema 7: Tipos de triângulos

Leia 3 valores de ponto flutuante, **A**, **B** e **C**, em ordem decrescente de valores, de modo que o lado **A** representa o maior dos três lados. Crie uma função para determina o tipo de triângulo que estes três lados formam, com base nos seguintes casos, sempre escrevendo uma mensagem adequada:

* Se A ≥ B + C, apresente a mensagem: NAO FORMA TRIANGULO;
* Se A² = B² + C² , apresente a mensagem: TRIANGULO RETANGULO;
* Se A² > B² + C² , apresente a mensagem: TRIANGULO OBTUSANGULO;
* Se A² < B² + C² , apresente a mensagem: TRIANGULO ACUTANGULO;
* Se os três lados forem iguais, apresente a mensagem: TRIANGULO
EQUILATERO;
* Se apenas dois dos lados forem iguais, apresente a mensagem: TRIANGULO
ISOSCELES.

### Entrada

A entrada contem três valores de ponto flutuante de dupla precisão: A (0 < A), B (0 < B) e C (0 < C).

### Saída

Imprima todas as classificações do triângulo especificado na entrada.

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
|7.0<br>7.0<br>5.0 | TRIANGULO ACUTANGULO<br>TRIANGULOS ISOSCELES|

|7.0<br>5.0<br>2.0 | NAO FORMA TRIANGULO |
|10.0<br>8.0<br>6.0| TRIANGULO RETANGULO|

```
# SP
def triangulo(A,B,C):
  if A >= B+C:
    print("Nao forma triangulo")
  else:
    if A**2 > B**2 + C**2:
      print("Triangulo Obstusangulo")
    elif A**2 < B**2 + C**2:
      print("Triangulo Acutangulo")
    else:
      print("Triangulo Retangulo")

    if  A==B and B==C and A==C:
      print("Triangulo Equilatero")
    elif A==B or B==C or A==C:
      print("Tringulo Isosceles")
  return None
```
