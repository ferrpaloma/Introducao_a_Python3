## Problema 1 : Maior valor

Receba varios valores em uma linha e crie uma lista. Crie uma função chamada **maximo** que receba a lista de valores e retorne o maior valor da lista para o programa principal. Todas as entradas tem que ser fornecidas pelo usuário

### Entrada

A entrada contém uma lista de valores separada por espaços em branco

### Saída

Imprima o maior valor da lista retornada pela função

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| 4 7 6 3 8 4 | maior = 8 |
| 0 5 3 4 1 4 | maior = 5 |
| 3 8 7 2 9 4 | maior = 9 |
| 7 6 6 8 3 2 | maior = 8 |

```
# SP
def maximo(lista):
  maior = float(lista[0])
  for valor in lista[1:]:
    if float(valor) > maior:
      maior = float(valor)
  return maior
```

## Problema 3 : Média

Receba varios valores em uma linha e crie uma lista. Crie uma função chamada **media** que receba uma lista de valores e retorne a média dos valores da lista para o programa principal. Todas as entradas tem que ser fornecidas pelo usuário

**Não usar**: sum() e len() (para o calculo da media)

### Entrada

A entrada contém uma lista de valores separada por espaços em branco

### Saída

Imprima a média dos valores da lista retornada pela função com 3 casas decimais

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| 2 17 9 1 15 4 18 | media = 9.429 |
| 7 3 9 7 | menor = 6.500 |
| 13 6 49 21 22 34 26 3 23 32 | menor = 22.900 |

```
# SP
def calculo_media(lista):
  soma = 0
  qtd = 0
  for indice in range(len(lista)):
    soma += float(lista[indice])
    qtd += 1
  return soma/qtd  
```

```
lista = input("Insira uma lista de valores espacados por um espaco em branco: ").split(sep = " ")
media = calculo_media(lista)
print("media = %.3f" % (media))
```

```
lista = input("Insira uma lista de valores espacados por um espaco em branco: ").split(sep = " ")
media = calculo_media(lista)
print("media = %.3f" % (media))
```

```
lista = input("Insira uma lista de valores espacados por um espaco em branco: ").split(sep = " ")
media = calculo_media(lista)
print("media = %.3f" % (media))
```

```
def transformar(lista):
  for indice in range(len(lista)):
    lista[indice] = float(lista[indice])
  return lista
```

```
lista = input("Insira uma lista de valores espacados por um espaco em branco: ").split(sep = " ")
lista = transformar(lista)
# map(int,lista)
media = sum(lista)/len(lista)
print("media = %.3f" % (media))
```

## Problema 4 : Pares e impares

Receba varios valores em uma linha e crie uma lista. Crie uma função chamada **pares_impares** que receba uma lista de valores e retorne a lista de valores pares e uma outra lista de valores impares para o programa principal. Todas as entradas tem que ser fornecidas pelo usuário

### Entrada

A entrada contém uma lista de valores separados por espaços em branco

### Saída

Imprima a lista de valores pares e a lista de valores impares

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| 93 99 78 52 59 92 15 83 70 79<br>42 39 76 19 40 6 57 20 45 62<br>83 32 35 58 35 34 63 71 71 76<br>90 47 62 60 24 44 88 93 70 6<br>60 72 78 4 27 49 61 56 77 13 | pares = [78.0, 52.0, 92.0, 70.0, 42.0, 76.0, 40.0, 6.0, 20.0, 62.0,<br>32.0, 58.0, 34.0, 76.0, 90.0, 62.0, 60.0, 24.0, 44.0, 88.0,<br>70.0, 6.0, 60.0, 72.0, 78.0, 4.0, 56.0]<br><br>impares = [93.0, 99.0, 59.0, 15.0, 83.0, 79.0, 39.0, 19.0, 57.0, 45.0,<br>83.0, 35.0, 35.0, 63.0, 71.0, 71.0, 47.0, 93.0, 27.0, 49.0,<br>61.0, 77.0, 13.0] |

```
# SP
def pares_impares(lista):
  pares = []
  impares = []
  for valor in lista:
    if valor%2 == 0:
      pares.append(valor)
    else:
      impares.append(valor)
  return (pares,impares)
```

```
lista = input("Insira uma lista de valores espacados por um espaco em branco: ").split(sep = " ")
lista = list(map(float,lista)) # transforma os valores em float
(pares,impares) = pares_impares(lista)
print("pares =",pares)
print("impares =",impares)
```

## Problema 5: Inverso

Receba varios valores em uma linha e crie uma lista. Crie uma função chamada **inverter** que receba uma lista de valores e retorne a ordem inversa da lista. Todas as entradas tem que ser fornecidas pelo usuário

**Não usar**: reverse()

### Entrada

A entrada contém uma lista de valores separada por espaços em branco

### Saída

Imprima a lista inversa

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| 2 17 9 1 15 4 18 | [18, 4, 15, 1, 9, 17, 2] |
| 7 3 9 7 | [7, 9, 3, 7] |
| 13 6 49 21 22 34 26 3 23 32 | [32, 23, 3, 26, 34, 22, 21, 49, 6, 13] |

```
# SP
def inverter(lista):
  lista_inversa = []
  for valor in lista[-1:-len(lista)-1:-1]:
    lista_inversa.append(valor)
  return lista_inversa
```

```
lista = input("Insira uma lista de valores espacados por um espaco em branco: ").split(sep = " ")
lista_inversa = inverter(lista)
print(lista_inversa)
```

## Problema 6: Distância

Receba dois valores por linha, recebendo 5 linha de valores. Os dois valores por linha formarão tuplas que devem ser armazenadas em uma lista. Considere essas tuplas como as posições (x,y) de um ponto em um plano de duas dimensões.Crie uma função chamada **distancia** que receba a lista de tuplas e calcule a distancia dos pontos até o centro da plano (0,0). A funcao deve retorne a posição, na ordem de entrada, do ponto mais distante ao centro do plano (0,0). Todas as entradas tem que ser fornecidas pelo usuário


Distância ao centro do plano: $d_{p} = \sqrt{x^2+y^2}$

### Entrada

A entrada contém uma lista de valores separados por espaços em branco

### Saída

Imprima a lista de valores pares e a lista de valores impares

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
|  0 -9<br>-1 -3<br>3 8<br>1 -5<br>8 10 | posicao = 4 |
| -8 -5<br>9 2<br>1 8<br>-9 3<br>7 4 | posicao = 3 |

```
# SP
def distancia(lista):
  distancias = []
  for x,y in lista:
    dist = (x**2+y**2)**(1/2)
    distancias.append(dist)
  maior_distancia = maximo(distancias)
  posicao = distancias.index(maior_distancia)
  return posicao

def maximo(lista):
  maior = float(lista[0])
  for valor in lista[1:]:
    if float(valor) > maior:
      maior = float(valor)
  return maior
```

```
# PP
lista = []
for i in range(5):
  x,y = input("Insira os pontos x y: ").split(sep = " ")
  lista.append((float(x),float(y)))

posicao = distancia(lista)
print("posicao =",posicao)
```

## Problema 7: Ordem decrescente

Receba varios valores em uma linha e crie uma lista. Crie uma função chamada **decrescente** que receba uma lista de valores e retorne a lista de valores em ordem descrescente para a função principal. Todas as entradas tem que ser fornecidas pelo usuário

**OBS1**: Não devem usar sorted()

**OBS2**: Usem o metodos disponiveis

### Entrada

A entrada contém uma lista de valores separados por espaços em branco

### Saída

Imprima a lista de valores em ordem descrecente

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| 3 29 17 10 27 6 8 26 1 0  | 29 27 26 17 10 8 6 3 1 0 |
| 1 21 20 4 25 9 23 17 1 2  | 25 23 21 20 17 9 4 2 1 1 |

```
help(list.remove)
```

```
# SP
def descrescente(lista):
  lista_decrescente = []
  for i in range(len(lista)):
    maior_valor = maximo(lista)
    lista.remove(maior_valor)
    lista_decrescente.append(maior_valor)
  return lista_decrescente

def maximo(lista):
  maior = float(lista[0])
  for valor in lista[1:]:
    if float(valor) > maior:
      maior = float(valor)
  return maior
```

```
# PP
lista = input("Insira uma lista de valores espacados por um espaco em branco: ").split(sep = " ")
lista = list(map(float,lista)) # transforma os valores em float
lista_decrescente = descrescente(lista)
print(lista_decrescente)
```

```
# SP
def decrescente(lista):
  lista_decrescente = []
  while lista != []:
    indice_maior = posicao_maior(lista)
    maior_valor = lista.pop(indice_maior)
    lista_decrescente.append(maior_valor)
  return lista_decrescente

def posicao_maior(lista):
  maior_valor = lista[0]
  posicao_maior_valor = 0
  for indice, valor in enumerate(lista):
    if valor > maior_valor:
      maior_valor = valor
      posicao_maior_valor = indice
  return posicao_maior_valor
```
