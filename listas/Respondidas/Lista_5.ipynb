## Problema 1: Encontre o N

Em um arquivo texto chamado "**encontre_N.txt**" temos um arquivo com diversas letras "**M**" e com algumas letras "**N**", processe o arquivo e encontre a fileira e a coluna (x,y) de cada uma das letras "**N**" dentro do arquivo. Os espaços em branco não são para ser considerados colunas. No final guarde as posições de **N** em um arquivo chamado "**posicoes_N.txt**", uma posição por linha. Para isto faça:

 1. crie uma função chamada "**carregar_arquivos**" que recebe o caminho do arquivo texto ("**encontre_N.txt**"), carregue todo o texto em uma lista e retorne essa lista para o programa principal.
 2. crie uma função chamada "**encontrar_Ns**" que recebe a lista de dados, que encontre a fileira e a coluna de cada "**N**" e as retorne em uma lista.
 3. crie uma função chamada "**guardar_arquivos**" que recebe o caminho do arquivo texto ("**posicoes_N.txt**") e a lista de posições e salve a lista no arquivo texto.

```
# SP
def carregar_arquivos(nome):
  arquivo = open(caminho+nome,"r")
  dados = []
  for linha in arquivo:
    linha = linha.strip() # limpeza na linha
    linha = linha.split() # divido em uma lista
    dados.append(linha)
  arquivo.close()
  return dados

def encontrar_Ns(dados):
  posicoes_Ns = []
  for i in range(len(dados)): # Linhas de dados
    for j in range(len(dados[i])): # Colunas de dados
      if dados[i][j] == "N":
        posicoes_Ns.append((i+1,j+1))
  return posicoes_Ns

def guardar_arquivos(posicoes_Ns,nome):
  arquivo = open(caminho+nome,"w")
  for i,j in posicoes_Ns:
    arquivo.write("{} {}\n".format(i,j))
  arquivo.close() 
  return None
```

```
# PP
nome = input("Digite o nome do arquivo com a extensão: ")
dados = carregar_arquivos(nome)
posicoes_Ns = encontrar_Ns(dados)
nome = input("Digite o nome do arquivo com a extensão: ")
guardar_arquivos(posicoes_Ns,nome)
```

## Problema 2: Arquivo de nomes

Receba do usuário repetido nomes de pessoas até que a o nome "**pare**" seja digitado, todos os nomes devem ser armazenados em um arquivo texto e imprima a média do tamanho dos nomes digitados. 

### Entrada

A entrada contém os nome digitado pelos usuário e um nome de um arquivo texto


```
# SP
def receber_nomes():
  nomes = []
  while True:
    nome = input("Digite um nome: ")

    if nome == "pare":
      print("O programa parou")
      break
    
    nomes.append(nome)
  return nomes

def media_tamanho_nomes(nomes):
  soma = 0
  for nome in nomes:
    soma += len(nome)
  media = soma/len(nomes)
  return media

def guardar_nomes(nomes):
  nome_arquivo = input("Digite o nome do arquivo com a extensão: ")
  arquivo = open(caminho+nome_arquivo,"w")
  for nome in nomes:
    arquivo.write("{}\n".format(nome))
  arquivo.close()
  return None
```

```
# PP
nomes = receber_nomes()
media = media_tamanho_nomes(nomes)
guardar_nomes(nomes)
print("A media de nomes e: {0:.3f}".format(media)) 
```


## Problema 3: Organizar nomes

Peça ao usuário o nome de dois arquivos texto (**Nomes1.txt**, **Nomes2.txt**) contendo uma lista de nomes (cada nome em uma linha). Processe os arquivos e em seguida, peça o nome de um terceiro arquivo que receberá o conteúdo dos dois arquivos anteriores alfabeticamente ordenado. Considere que o conteúdo dos arquivos de entrada não está ordenado inicialmente.

**OBS**: Pesquise sobre a função sorted()

### Entrada

A entrada contém o nome dos dois arquivo texto e o nome de um terceiro para armazenar


```
# SP
def receber_nomes_2(nome_do_arquivo_1,nome_do_arquivo_2):
  nomes = []

  # Receber nomes do primeiro arquivo
  arquivo = open(caminho+nome_do_arquivo_1,"r")
  for linha in arquivo:
    linha = linha.strip() # limpeza na linha
    nomes.append(linha)
  arquivo.close()

  # Receber nomes do segundo arquivo
  arquivo = open(caminho+nome_do_arquivo_2,"r")
  for linha in arquivo:
    linha = linha.strip() # limpeza na linha
    nomes.append(linha)
  arquivo.close()

  return nomes

def guardar_nomes_ordenados(nomes):
  nome_arquivo = input("Digite o nome do arquivo com a extensão: ")
  arquivo = open(caminho+nome_arquivo,"w")
  for nome in sorted(nomes):
    arquivo.write("{}\n".format(nome))
  arquivo.close()
  return None
```

```
# PP
nome_do_arquivo_1 = input("Digite o nome do primeiro arquivo com a extensão: ")
nome_do_arquivo_2 = input("Digite o nome do segundo arquivo com a extensão: ")
nomes = receber_nomes_2(nome_do_arquivo_1,nome_do_arquivo_2)
guardar_nomes_ordenados(nomes)
```

## Problema 4: Menor e maior comprimento

Processe um arquivo texto cujo nome seja escolhido pelo usuário e escreva na tela as posições das linhas que contenham as cadeias de caracteres (strings) de menor e maior comprimento. Caso o arquivo esteja vazio, escreva a seguinte mensagem na saída padrão: "**O arquivo está vazio**".

### Entrada

A entrada consiste do nome do arquivo

### Saída

A saída consiste da posição das linhas que contém a menor e a maior quantidade de caracteres, respectivamente (considere a quantidade de espaços em branco na contagem). Caso haja empate (seja para menor ou maior quantidade de caracteres) exiba a posição da última linha lida que possua aquele valor. Os valores devem ser exibidos na mesma linha, separados por um espaço em branco.

OBS: Não esqueça de eliminar os caracteres especiais que são invisíveis no texto usando o "strip()"

| **Exemplos de Entrada** | **Exemplos de Saída** |
|:-|:-|
| tarantino.txt | 5 4 |
| burton.txt | 5 6 |
| spielberg.txt | O arquivo esta vazio |


```
# SP
def receber_filmes(nome_do_arquivo):
  arquivo = open(caminho+nome_do_arquivo,"r")
  filmes = []
  for linha in arquivo:
    linha = linha.strip() # limpeza na linha
    filmes.append(linha)
  arquivo.close()
  return filmes

def maior_linha(filmes):
  maior = filmes[0]
  for filme in filmes[1:]:
    if len(filme) >= len(maior):
      maior = filme
  pos_maior = filmes.index(maior)
  return pos_maior+1

def menor_linha(filmes):
  menor = filmes[0]
  for filme in filmes[1:]:
    if len(filme) <= len(menor):
      menor = filme
  pos_menor = filmes.index(menor)
  return pos_menor+1
```

```
# PP
nome_do_arquivo = input("Digite o nome do arquivo com a extensão: ")
filmes = receber_filmes(nome_do_arquivo)
if filmes != []:
  pos_menor = menor_linha(filmes)
  pos_maior = maior_linha(filmes)
  print("{} {}".format(pos_menor,pos_maior))
else:
  print("O arquivo esta vazio")
```

## Problema 5: 

Peça ao usário o nome de um arquivo texto ("**Notas.txt**") contendo nome de alunos e suas notas espaçado por um espaço em branco e processe o arquivo texto. Substitue a nota de todos os alunos com a inicial do nome maior ou igual a "P" para 10 e escreva em um arquivo texto.

### Entrada

A entrada contém o nome de um arquivo texto

```
def carregar_arquivos(nome):
  arquivo = open(caminho+nome,"r")
  dados = []
  for linha in arquivo:
    linha = linha.strip() # limpeza na linha
    linha = linha.split() # divido em uma lista
    dados.append(linha)
  arquivo.close()
  return dados

def substituir(nome_notas):
  for pos in range(len(nome_notas)):
    if nome_notas[pos][0] >= "P":
      nome_notas[pos][1] = 10
  return nome_notas

def guardar_arquivos(nome_notas):
  nome_do_arquivo = input("Digite o nome do arquivo com a extensão: ")
  arquivo = open(caminho+nome_do_arquivo,"w")
  for i,j in nome_notas:
    arquivo.write("{} {}\n".format(i,j))
  arquivo.close() 
  return None
```

```
# PP
nome_do_arquivo = input("Digite o nome do arquivo com a extensão: ")
nome_notas = carregar_arquivos(nome_do_arquivo)
nome_notas = substituir(nome_notas)
guardar_arquivos(nome_notas)
```


## Problema 6: Pontos dentro de uma circunferência

Peça ao usário o nome de um arquivo ("**pontos.txt**") contendo um ponto com coordenadas x e y espaçados por um espaço em branco (um ponto por linha) e processe o arquivo texto. Peça também ao usúario um número inteiro **n** (suponha que 0 < **n** <= 5), sendo **n** o raio de uma circunferência com origem no centro do plano. Encontre os pontos que estejam fora da circunferência e salve em um arquivo texto os pontos fora da circunferência

Exemmplo do nome do arquivo: pontos_raio_n.txt (**n** varia com o **n** que você decedir).

**OBS**: Para um ponto esteja fora de um circunferência, a distancia do ponto ao centro tem que ser maior que o raio.

Distância ao centro do ponto: $d_{p} = \sqrt{x^2+y^2}$


### Entrada

A entrada contém o nome do arquivo texto e um número inteiro que corresponde ao raio da circunferência

```
def carregar_arquivos(nome):
  arquivo = open(caminho+nome+".txt","r")
  dados = []
  for linha in arquivo:
    linha = linha.strip() # limpeza na linha
    linha = linha.split() # divido em uma lista
    dados.append(linha)
  arquivo.close()
  return dados

def distancia(lista):
  distancias = []
  for x,y in lista:
    dist = (int(x)**2+int(y)**2)**(1/2)
    distancias.append(dist)
  return distancias

def guardar_pontos(pontos,distancias,raio):
  nome_do_arquivo = input("Digite o nome do arquivo sem a extensão: ")
  arquivo = open(caminho+nome_do_arquivo+"_{}".format(raio)+".txt","w")
  for pos in range(len(pontos)):
    if distancias[pos] > raio:
      arquivo.write("{} {}\n".format(pontos[pos][0],pontos[pos][1]))
  arquivo.close() 
  return None
```

```
raio = int(input("Digite o raio da circunferencia: "))
nome_do_arquivo = input("Digite o nome do arquivo sem a extensão: ")
pontos = carregar_arquivos(nome_do_arquivo)
distancias = distancia(pontos)
guardar_pontos(pontos,distancias,raio)
```
